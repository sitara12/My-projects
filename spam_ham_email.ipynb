```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Step 1: Collect Data Set
# Assuming you have uploaded a CSV file named 'emails.csv' with 'text' and 'label' columns representing email messages and their labels respectively
from google.colab import files
uploaded = files.upload()

# Load the CSV file into a Pandas DataFrame
# Load the CSV file into a Pandas DataFrame
dataset = pd.read_csv('emails.csv')


# Step 2: Pre-processing
# Assuming you have already performed data cleaning and handling missing values
# Noisy values can be addressed through techniques like stemming or lemmatization

# Step 3: Feature Selection
stop_words = set(stopwords.words('english'))

# Tokenize and remove stop words from text
dataset['processed_text'] = dataset['text'].apply(lambda x: ' '.join([word for word in word_tokenize(str(x).lower()) if word.isalpha() and word not in stop_words]))

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(dataset['processed_text'])
# Use the processed_text column
y = dataset['label']

# Step 4: Apply Spam Filter Algorithms
# Step 4.1: Handle Data - Split into training and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4.2: Train the Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

# Step 4.3: Make Predictions
predictions = classifier.predict(X_test)

# Step 4.4: Evaluate Accuracy
accuracy = metrics.accuracy_score(y_test, predictions)

# Step 5: Train & Test Data (already done in Step 4.1)

# Step 6: Confusion Matrix
confusion_matrix = metrics.confusion_matrix(y_test, predictions)

# Step 7: Accuracy
# Assuming you want to compare the accuracy of Naive Bayes with another algorithm, e.g., Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_predictions = logreg.predict(X_test)
logreg_accuracy = metrics.accuracy_score(y_test, logreg_predictions)
# Print the results
print("Naive Bayes Accuracy:", accuracy)
print("Logistic Regression Accuracy:", logreg_accuracy)
print("\nConfusion Matrix:")
print(confusion_matrix)






```
